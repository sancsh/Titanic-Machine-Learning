```python
x_train, x_cv, y_train, y_cv = train_test_split(train, y, test_size =0.25)
```

```python
x_train = np.asmatrix(x_train)
x_cv = np.asmatrix(x_cv)
```

```python
y_train = np.reshape(y_train, (666,1))
y_cv = np.reshape(y_cv,(223,1))
```

```python
x_train.std(axis=0)
```

```python
x_train = (x_train - x_train.mean(axis=0)) / x_train.std(axis=0)
```

```python
def sigmoid(z):
    return 1.0/(1+ np.exp(-z))
```

```python
x_train= np.insert(x_train,0,values= 1, axis=1)

m = y_train.shape[0]
```

```python
def cost_function(theta, x, y, alpha =0.003):
    J=0
    grad = np.zeros(theta.shape)
    J = (1/m) * np.sum(np.dot(np.transpose(-y),np.log(sigmoid(np.dot(x, theta)))) - np.dot(np.transpose(1-y), np.log(1-sigmoid(np.dot(x,theta)))))
    grad = (alpha/m) * (np.dot(np.transpose(x),sigmoid(np.dot(x,theta)) - y))
    return (J, grad)
    
```

```python
theta = np.zeros((x_train.shape[1],1))
```

```python
for i in range(0,500):
    (J, grad)= cost_function(theta, x_train, y_train)
    theta = theta - grad
```

```python
def predict(z):
    if z>= 0.5:
        return 1
    else:
        return 0
    
```

```python
x_cv = np.insert(x_cv,0,1,axis=1)
```

```python
hyp = sigmoid(np.dot(x_cv, theta))
```

```python
pred = np.apply_along_axis(predict,axis=1,arr= hyp)
```
